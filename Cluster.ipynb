#### Reading the table from the given link

```python
#Import requied packages
import pandas as pd
import numpy as np
```

```python
#Reading data as a dataframe
url='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
dataframe = pd.read_html(url)[0]
dataframe.head()
```

```python
#Dropping the rows with Borough is Not assigned
dataframe['Borough'].replace("Not assigned", np.nan, inplace=True)
dataframe.dropna(inplace=True)#Drop the missing data
```

```python
#Reset the index
dataframe.reset_index(drop=True, inplace=True)
dataframe.head()
```

```python
dataframe.shape#Shape of the dataframe
```

```python
#Reading lat long data
url = 'https://cocl.us/Geospatial_data'
lat_long_df = pd.read_csv(url)
lat_long_df.head()#Displaying top 5 rows
```

```python
neighborhoods = pd.merge(dataframe, lat_long_df, on='Postal Code')
```

```python
import numpy as np # library to handle data in a vectorized manner

import pandas as pd # library for data analsysis
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

import json # library to handle JSON files

from geopy.geocoders import Nominatim # convert an address into latitude and longitude values

import requests # library to handle requests
from pandas.io.json import json_normalize # tranform JSON file into a pandas dataframe

# Matplotlib and associated plotting modules
import matplotlib.cm as cm
import matplotlib.colors as colors

# import k-means from clustering stage
from sklearn.cluster import KMeans

#!conda install -c conda-forge folium=0.5.0 --yes # uncomment this line if you haven't completed the Foursquare API lab
import folium # map rendering library

print('Libraries imported.')
```

```python
print('The dataframe has {} boroughs and {} neighborhoods.'.format(
        len(neighborhoods['Borough'].unique()),
        neighborhoods.shape[0]
    )
)
```

```python
address = 'Toronto'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
# create map of New York using latitude and longitude values
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(neighborhoods['Latitude'], neighborhoods['Longitude'], neighborhoods['Borough'], neighborhoods['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
